```python
# 1. 일별수익률추이
def daily_ret(result):
    report = result.copy()
    report['일별수익률'] = result['port_val'].pct_change()
    report['누적수익률'] = result['port_val'] / result['port_val'][0] - 1
    report.replace(np.nan, 0, inplace=True)
    del report['strategy']
    report1 = report
    return report1
```
